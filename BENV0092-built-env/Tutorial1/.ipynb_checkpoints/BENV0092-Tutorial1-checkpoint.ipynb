```r
#ENERGY_CONSUMPTION_CURRENT is in KWH/m2 and it is the sum of electricity and gas consumption for each home
#create a new variable with consumption/m2 * m2, to get consumption.
EPC_tidy$ENERGY_CONSUMPTION <- EPC_tidy$ENERGY_CONSUMPTION_CURRENT * EPC_tidy$TOTAL_FLOOR_AREA
summary(EPC_tidy$ENERGY_CONSUMPTION)
```

```r
#######lets see the distribution of consumption
g1 <- ggplot(EPC_tidy, aes(x = ENERGY_CONSUMPTION))+geom_freqpoly(aes(y = ..count..)) +
  geom_vline(aes(xintercept = mean(ENERGY_CONSUMPTION)), 
             linetype = "dashed")+ xlim(0, 200000) #'aes()': a connection between the data (the dataframe) 
#and the visual objects that we draw (the geoms). 
#the aes() function allows you to map the variables in your data frame to the aesthetic attributes of the 
#geometric objects of your plot
g1

#Try changing the line type to 'solid' and specify a color to the vertical line and see how this looks!
#http://www.sthda.com/english/wiki/ggplot2-line-types-how-to-change-line-types-of-a-graph-in-r-software 

#my Median is 17569 while my mean is 59690 - pretty far apart! I have outliers!
#What is an outlier? Defined as an observation which stands far away from the most of other observations. 
#Often a outlier is present due to a measurement error. 
#Therefore, one of the most important tasks in data analysis is to identify and (if is necessary) to remove the outliers.
```

```r
#######Lets omit our outliers
#Adan and Fuerst (2016) and others: valid records defined as properties with 
#electricity consumption values between 100 and 25,000 kWh/yr 
#and Gas consumption values between 3000 and 50,000 kWh/yr

EPC_tidy <-filter(EPC_tidy, ENERGY_CONSUMPTION > 3100 & ENERGY_CONSUMPTION < 75000)

#How can I check how many households I have now? we are left with 65911 households
dim(EPC_tidy)

#how does the frequency polygon look now?
g2 <- ggplot(EPC_tidy, aes(x = ENERGY_CONSUMPTION))+geom_freqpoly(aes(y = ..count..)) +
  geom_vline(aes(xintercept = mean(ENERGY_CONSUMPTION)), 
             linetype = "dashed")
g2
summary(EPC_tidy$ENERGY_CONSUMPTION)#much better
```

```r
###### Regression Model Fitting 
#I will only keep variables that I will need in my analysis:
EPC_tidy<-select(EPC_tidy,c("ENERGY_CONSUMPTION","TOTAL_FLOOR_AREA", "WALLS_ENERGY_EFF","ROOF_ENERGY_EFF","MAINHEAT_ENERGY_EFF","LIGHTING_ENERGY_EFF"))
#I will exclude levels of factors coded as N/A 

EPC_tidy<-filter(EPC_tidy, WALLS_ENERGY_EFF!="N/A"&ROOF_ENERGY_EFF!="N/A"&MAINHEAT_ENERGY_EFF!="N/A"&LIGHTING_ENERGY_EFF!="N/A")

#Lets start with one variable and we will be adding the rest one-by-one. 
#We start with the variable that will most probably be a very good predictor of consumption
fit1<- lm(ENERGY_CONSUMPTION ~   TOTAL_FLOOR_AREA, data=EPC_tidy)
summary(fit1)
#R-squared:  0.46, What does it mean?
#Means that a household's energy consumption is explained by its floor area by 46%. 
#The remaining 54% of the variation is explained by other factors.
#The Estimate of TOTAL FLOOR AREA represents the difference in the predicted consumption for each one-unit difference in TOTAL FLOOR AREA 
#For a given predictor, the t-statistic evaluates whether or not there is significant association between the predictor and the outcome variable, 
#that is whether the beta coefficient of the predictor is significantly different from zero

#lets look at our P Value: using TOTAL_FLOOR_AREA gives a highly significant result (p<0.05). #The P-value is the probability that 
#a t statistic having n-2 degrees of freedom is more extreme than my t statistic: 
#a low p-value means a low chance of the null hypothesis being true. Here, the null hypothesis would be that the 2 variables are NOT correlated. 
#We should reject the NULL hypothesis.

#So we have TOTAL FLOOR AREA and number of heated rooms. these two variables essentially refer to almost the same thing: how big the house is.
#we cannot include them both in the model: multicollinearity problem!! we need to be careful
#but we can form 2 seperate models for these 2 variables and see which one explains more variation in consumption
#I have already done this and found out that Total floor area is a much better predictor than the Number of Heated rooms.

#we continue by adding more variables one-by-one
 
fit2<- lm(ENERGY_CONSUMPTION ~  TOTAL_FLOOR_AREA+ WALLS_ENERGY_EFF, data=EPC_tidy)

summary(fit2)
```

```r
#lets see our Estimates for WALLS_ENERGY_EFF, which is a Categorical variable:
#These Estimates are the difference in consumption for each one-unit difference in WALLS_ENERGY_EFF (Switching from category Average to another) 
#if TOTAL_FLOOR_AREA remains constant
#In other words, for example, the Estimate for WALLS_ENERGY_EFFGood is the difference between the mean of WALLS_ENERGY_EFFGood and 
#the mean of WALLS_ENERGY_EFFAverage (our reference)
#Here this is negative, showing us that on average, mean consumption of households with Good Wall Energy Efficiency will have lower consumption than 
#households with an Average Wall Energy Efficiency

anova(fit1,fit2)#RSS Decreased 
#we want RSS (Residual Sum of Squares) to decrease as we are adding more variables - it tells us that the degree of inacurracy decreases.
#RSS: tells us about the variability that cannot be explained by our model: obviously, we want this to be as small as it can be
#The F-Statistic: Variation Between Sample Means / Variation Within the Samples: We don't care too much about how to interprete this. 

fit3<- lm(ENERGY_CONSUMPTION ~  TOTAL_FLOOR_AREA + WALLS_ENERGY_EFF+ ROOF_ENERGY_EFF, data=EPC_tidy)
summary(fit3)
```

```r
anova(fit2,fit3)
fit4<- lm(ENERGY_CONSUMPTION ~  TOTAL_FLOOR_AREA + WALLS_ENERGY_EFF+ ROOF_ENERGY_EFF+LIGHTING_ENERGY_EFF, data=EPC_tidy)
summary(fit4)
#better to look at R Squared Adjusted as we add more variables. a modified version of R-squared that has been adjusted for the number of predictors in the model. 
#The adjusted R-squared increases only if the new term improves the model more than would be expected by chance. 
#It decreases when a predictor improves the model by less than expected by chance.
anova(fit3,fit4)#only a slightly better fit by adding LIGHTING_ENERGY_EFF

```
